```python
import rasterio

mars = rasterio.open('../../Planets/mars/data/Mars_MGS_MOLA_DEM_mosaic_global_463m.tif')
mars = mars.read()
```

```python
import numpy as np
print(mars.shape)
print(np.amin(mars[0]))
print(np.amax(mars[0]))
print(np.amax(mars[0]) + abs(np.amin(mars[0])))
```

```python
from matplotlib.colors import BoundaryNorm, LinearSegmentedColormap
import numpy as np

custom_cmap = LinearSegmentedColormap.from_list('mars', ['#162252',
                                                         '#104E8B',
                                                         '#00B2EE',
                                                         '#00FF00', 
                                                         '#FFFF00',
                                                         '#FFA500',
                                                         '#FF0000', 
                                                         '#8b0000',
                                                         '#964B00', 
                                                         '#808080',
                                                         '#FFFFFF'], N=2221)

bounds = np.arange(-8210, 14000, 10)
norm = BoundaryNorm(bounds, custom_cmap.N)
```

```python
colors_undersea = plt.cm.ocean(np.linspace(0.2, 1, 821))
undersea_map = LinearSegmentedColormap.from_list('undersea_map', colors_undersea, N=821)

colors_land = plt.cm.terrain(np.linspace(0.25, 1, 1400))
land_map = LinearSegmentedColormap.from_list('land_map', colors_land, N=1400)

colors = np.vstack((colors_undersea, colors_land))
terrain_map = LinearSegmentedColormap.from_list('cut_terrain', colors, N=2221)

bounds = np.arange(-8210, 14000, 10)
norm = BoundaryNorm(bounds, terrain_map.N)
```

```python

below_sea_level = np.arange(np.amin(mars[0]), 0, 1)
above_sea_level = np.arange(0, np.amax(mars[0]), 1)
all_values = np.arange(np.amin(mars[0]), np.amax(mars[0]), 1)

below_sea_level_gradient = np.vstack((below_sea_level, below_sea_level))
above_sea_level_gradient = np.vstack((above_sea_level, above_sea_level))
all_values_gradient = np.vstack((all_values, all_values))

fig = plt.figure()
fig.set_size_inches(15, 11)

ax1 = plt.subplot(3,1,1)
ax1.imshow(below_sea_level_gradient, aspect='auto', cmap=undersea_map)
ax1.get_yaxis().set_visible(False)
ax1.set_xlabel("Altitude", fontsize=13)
ax1.set_xticks(ticks=np.arange(0, 8000, 1000))
ax1.set_xticklabels(labels=np.arange(-8000, 0, 1000))
ax1.set_title("Ocean cmap", fontsize=13)
ax1.tick_params(labelsize=13)

ax2 = plt.subplot(3,1,2)
ax2.imshow(above_sea_level_gradient, aspect='auto', cmap=land_map)
ax2.get_yaxis().set_visible(False)
ax2.set_xlabel("Altitude", fontsize=13)
ax2.set_xticks(ticks=np.arange(0, 24000, 3000))
ax2.set_xticklabels(labels=np.arange(0, 24000, 3000))
ax2.set_title("Terrain cmap", fontsize=13)
ax2.tick_params(labelsize=13)

ax3 = plt.subplot(3,1,3)
ax3.imshow(all_values_gradient, aspect='auto', cmap=terrain_map, norm=norm)
ax3.text(201, 0, "Hellas Basin", color='white', fontsize=13)
ax3.text(26500, 0, "Olympus Mons", fontsize=13)
ax3.text(2000, 1, "Vastitas Borealis", color='white', fontsize=13)
ax3.text(16000, 1, "Tharsis rise", fontsize=13)
ax3.text(501, 0.5, "Valles Marineris", color='white', fontsize=13)
ax3.text(23500, 0.5, "Ascreaus Mons", fontsize=13)
ax3.text(19500, 1, "Pavonis Mons", fontsize=13)
ax3.text(22500, 1.2, "Arsia Mons", fontsize=13)
ax3.text(5000, 1.2, "Argyre", fontsize=13)
ax3.text(5000, 0.5, "Utopia Planitia", fontsize=13)
ax3.text(12500, 0, "Elysium", fontsize=13)
ax3.set_xlabel("Altitude", fontsize=13)
ax3.set_title("Custom cmap", fontsize=13)
ax3.tick_params(labelsize=13)

ax3.axvline(0+8000, color='black')
ax3.set_xticks(ticks=np.arange(0, 30000, 4000))
ax3.set_xticklabels(labels=np.arange(-8000, 22000, 4000))
ax3.get_yaxis().set_visible(False)
plt.tight_layout()
plt.savefig("Outputs/raw5.png", dpi=400, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(14, 7)

imgs = plt.imshow(mars[0], cmap=terrain_map, norm=norm)
ax = label_features_terrain(ax)
ax.axis('off')
newax = fig.add_axes([0.8, 0.78, 0.08, 0.08], anchor='NE')
newax.imshow(logo)
newax.axis('off')
txt = ax.text(0.02, 0.03, "Martian Topography \n@PythonMaps",
              size=8,
              color='black',
              transform = ax.transAxes)
plt.savefig("Outputs/raw7.png", dpi=400, bbox_inches='tight')

plt.show()
```

```python
def change_sea_level(new_sea_level):
    new_sea_level = new_sea_level / 10
    undersea = int(800 + new_sea_level)
    land = int(1400 - new_sea_level)
    colors_undersea = plt.cm.ocean(np.linspace(0.2, 0.8, undersea))
    undersea_map = LinearSegmentedColormap.from_list('undersea_map', colors_undersea, N=undersea)

    colors_land = plt.cm.terrain(np.linspace(0.25, 1, land))
    land_map = LinearSegmentedColormap.from_list('land_map', colors_land, N=land)

    colors = np.vstack((colors_undersea, colors_land))
    terrain_map = LinearSegmentedColormap.from_list('cut_terrain', colors, N=2221)

    bounds = np.arange(-8210, 14000, 10)
    norm = BoundaryNorm(bounds, terrain_map.N)
    return terrain_map, norm
```

```python
terrain_map0, norm0 = change_sea_level(new_sea_level = 0)
terrain_map3000, norm3000 = change_sea_level(new_sea_level = 3000)
terrain_map_3000, norm_3000 = change_sea_level(new_sea_level = -3000)

x = np.arange(np.amin(mars[0]), np.amax(mars[0]), 1)
gradient = np.vstack((x, x))

fig = plt.figure()
ax1 = plt.subplot(3,1,1)
fig.set_size_inches(15, 11)
ax1.imshow(gradient, aspect='auto', cmap=terrain_map0, norm=norm0)
ax1.axvline(0+8000, color='black')
ax1.set_xticks(ticks=np.arange(0, 30000, 4000))
ax1.set_xticklabels(labels=np.arange(-8000, 22000, 4000))
ax1.get_yaxis().set_visible(False)
ax1.set_xlabel("Altitude", fontsize=13)
ax1.set_title("Sea Level - 0m", fontsize=13)
ax1.tick_params(labelsize=13)
ax2 = plt.subplot(3,1,2)
ax2.imshow(gradient, aspect='auto', cmap=terrain_map3000, norm=norm3000)
ax2.axvline(3000+8000, color='black')
ax2.set_xticks(ticks=np.arange(0, 30000, 4000))
ax2.set_xticklabels(labels=np.arange(-8000, 22000, 4000))
ax2.get_yaxis().set_visible(False)
ax2.set_xlabel("Altitude", fontsize=13)
ax2.set_title("Sea Level - 3000m", fontsize=13)
ax2.tick_params(labelsize=13)
ax3 = plt.subplot(3,1,3)
ax3.imshow(gradient, aspect='auto', cmap=terrain_map_3000, norm=norm_3000)
ax3.axvline(-3000+8000, color='black')
ax3.set_xticks(ticks=np.arange(0, 30000, 4000))
ax3.set_xticklabels(labels=np.arange(-8000, 22000, 4000))
ax3.get_yaxis().set_visible(False)
ax3.set_xlabel("Altitude", fontsize=13)
ax3.set_title("Sea Level - -3000m", fontsize=13)
ax3.tick_params(labelsize=13)
plt.tight_layout()
plt.savefig("Outputs/raw8.png", dpi=400, bbox_inches='tight')

plt.show()
```
